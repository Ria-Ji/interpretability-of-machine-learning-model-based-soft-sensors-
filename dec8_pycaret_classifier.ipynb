## 1.pycaret

```python
# df = data.copy()
df = data[:].copy()
```

```python
df.tail(n=1)
```

```python
train = df.iloc[:40000, :]
test  = df.iloc[40000:, :]
```

```python
from pycaret.classification import *
```

```python
s = setup(data = train, 
          test_data = test, 
          target = 'y', 
          fold_strategy = 'timeseries', 
          fold = 5, 
          # transform_target = True, 
          session_id = 111)
```

```python
best = compare_models()
```

```python
et = create_model("ada")
```

```python
predictions = et.predict(test.loc[:, test.columns!="y"])
```
